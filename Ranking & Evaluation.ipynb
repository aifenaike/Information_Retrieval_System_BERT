# Ranking & Evaluation

```python
import pandas as pd
import numpy as np
test_corpus = pd.read_csv("testing_corpus.csv")
test_ranks = pd.read_csv("data/test_results.csv")
train_ranks = pd.read_csv("data/train_results.csv")
test_queries = pd.read_csv("test_queries_.csv")
```

```python
from gensim.models import Word2Vec
#load saved doc embeddings
test_embeddings = np.load('data/word2vec/test_embeddings.npy',allow_pickle=True)
w2v_model = Word2Vec.load("models/word2vec.model")
```

```python
#evaluate IR system on test set over a search space {k: 1 ~ 10}
from search_space_utils import *
from evaluation import Evaluate

query_ = test_queries
MAP=[]
for top_k in range(10):
    evaluation = Evaluate(query_,test_ranks,w2v_model,top_k=top_k+1)
    map_k = evaluation.mean_average_precision(test_corpus,test_embeddings)
    MAP.append(map_k)
    print()
```

```python
limits = [f'MAP@{k+1}' for k in range(10)]
import seaborn as sns
pal = sns.color_palette("Greens_d", len(limits))
hue_order = np.array(MAP).argsort()
sns.barplot(y=limits, x=MAP, palette=np.array(pal[:])[hue_order]).set(title='Mean Average Precision across top k searches')
```
